<h3> Import packages we need </h3>

```python
import pandas as pd
import numpy as np
import keras

from keras.models import Sequential
from keras.layers import Dense

from sklearn.model_selection import train_test_split
```

<h2> Part 2: Concrete Data </h2> 

<p> Retrieve the data from https://cocl.us/concrete_data 

    To recap, the predictors in the data of concrete strength include:

    Cement
    Blast Furnace Slag
    Fly Ash
    Water
    Superplasticizer
    Coarse Aggregate
    Fine Aggregate
</p>

```python
concrete_data = pd.read_csv('https://cocl.us/concrete_data')

concrete_data.head()
```

<p> Separating the data between the predictors and the target. This is need for a regression network. The predictors is the data we have and the target is what we are trying to find. </p>

```python
predictors = concrete_data.drop(columns=['Strength'])
target = concrete_data['Strength']

n_cols = predictors.shape[1]
```

<h1> Part 3: Neural Network Assignment </h1>

<p> Creating a Neural network with one hidden layer of 10 nodes, and a ReLU activation function. This network uses the adam optimizer and the mean squared error as the loss function. </p>

```python
def regression_model():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

model = regression_model()
```

<p>  Randomly split the data into a training and test sets by holding 30% of the data for testing. You can use the train_test_splithelper function from Scikit-learn. </p>

```python
predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, 
                                                                                test_size=0.3, random_state=71)
predictors_train.shape
```

<h3> Train the model on the training data using 50 epochs. </h3>

```python
model.fit(predictors_train, target_train, epochs=50)
```

<p>  Evaluate the model on the test data and compute the mean squared error between the predicted concrete strength and the actual concrete strength. You can use the mean_squared_error function from Scikit-learn. </p> 

```python
predictions_test = model.predict(predictors_test)
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(target_test, predictions_test)
print('Mean squared error: %.3f' % (mse))
```

<h1> A: Evaluate the model 50 times and create a list of 50 mean squared errors. </h1>

```python

def evaluate_model(model_func, predictors, targets, epochs=50):
    mean_squared_errors = []
    for i in range(50):

        model = model_func()
        predictors_train, predictors_test, target_train, target_test = train_test_split(predictors, target, 
                                                                                    test_size=0.3, random_state=71)
        model.fit(predictors_train, target_train, epochs=epochs, verbose=0)

        predictions_test = model.predict(predictors_test)
        mse = mean_squared_error(target_test, predictions_test)
        mean_squared_errors.append(mse)
        print('.', end='')
    print(' Finished evaluating!')
    
    return np.mean(mean_squared_errors), np.std(mean_squared_errors)
```

```python
mean_mse, std_mse = evaluate_model(regression_model, predictors, target)

print("Mean squared errors for 50 regression models: mean = %.3f, std dev = %.3f" %(mean_mse, std_mse))
```

<h1> B: Normalize the data by subtracting the mean from the individual predictors and dividing by the standard deviation. Then rerun the model. </h1>

```python
predictors_norm = (predictors - predictors.mean()) / predictors.std()
predictors_norm.head()
```

```python
mean_mse, std_mse = evaluate_model(regression_model, predictors_norm, target)

print("Mean squared errors for 50 regression models on normalized data: mean = %.3f, std dev = %.3f" % (mean_mse, std_mse))
```

<h1> C: Rerun the model with 100 epochs. </h1>

```python
mean_mse, std_mse = evaluate_model(regression_model, predictors_norm, target, epochs=100)

print("Mean squared errors for 50 regression models on normalized data, trained 100 epochs: mean = %.3f, std dev = %.3f" % (mean_mse, std_mse))
```

<h1> D: Rerun the model with three hidden layers, each of 10 nodes and ReLU activation function. </h1>

```python
def regression_model_D():
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=(n_cols,)))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model


mean_mse, std_mse = evaluate_model(regression_model_D, predictors_norm, target)

print("Mean squared errors for 50 regression models with 3 hidden layers: mean = %.3f, std dev = %.3f" % (mean_mse, std_mse))
```
